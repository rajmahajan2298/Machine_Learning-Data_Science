```python
ages = [23,24,32,45,12,43,67,45,32,56,120]

import numpy as np
print(np.mean(ages))
print(np.median(ages))
```

```python
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
```

```python
print(statistics.mode(ages))
```

```python
import seaborn as sns
sns.boxplot(ages,orient="h")
```

```python
# 5 Number Summary -

import numpy as np
q1,q3 = np.percentile(ages,[25,75])
print(q1,q3)
```

```python
# To check Outlier - 

iqr = q3-q1
lower_fence = q1 -1.5*(iqr)
higher_fence = q3 +1.5*(iqr)
print(lower_fence, higher_fence)
```

```python
# Measure of Dispersion -

statistics.variance(ages)

def variance(data):
    n = len(ages)
    mean = sum(data)/n
    deviation = [(x - mean)**2 for x in data]
    variance = sum(deviation)/(n-1)
    return variance

import math
math.sqrt(statistics.pvariance(ages))
```

```python
#Check Whether Distribution is normal or not -

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    stat.probplot(sample,dist='norm', plot=pylab)
    plt.subplot(1,2,2)
    plt.show()
```

```python
# Create Normal Distributed Data -

sample = np.random.normal(0.5,0.2,1000)
plot_data(sample)
```
